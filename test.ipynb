```python
from datetime import datetime
import pytz

def nyc_now() -> datetime:
    return datetime.now(pytz.timezone("UTC")).replace(tzinfo=None)
nyc_now()
```

```python
from pandas_market_calendars import get_calendar, date_range
import pandas as pd

nyse = get_calendar("NYSE")
start_time = "2025-07-23 20:01:00"
end_time = "2025-07-24 19:00:00"
start_date = pd.to_datetime(start_time).tz_localize(None)
end_date = pd.to_datetime(end_time).tz_localize(None)

sched = nyse.schedule(start_date=start_date, end_date=end_date, tz="UTC")
canon_idx = date_range(sched, frequency="1min", closed="both")
canon_idx = canon_idx.tz_localize(None)
```

```python
canon_idx = [idx for idx in canon_idx if idx >= pd.to_datetime(start_time)]
canon_idx = [idx for idx in canon_idx if idx <= pd.to_datetime(end_time)]
canon_idx
```
